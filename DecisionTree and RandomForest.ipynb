# Creating dummy variable and dropping one column to avoid dummy variable trap

```python
relevent_experience_dummies = pd.get_dummies(df5.relevent_experience, drop_first = True)
education_level_dummies = pd.get_dummies(df5.education_level, drop_first = True)
major_discipline_level_dummies = pd.get_dummies(df5.major_discipline, drop_first = True)
```

```python
df6 = pd.concat([df5,relevent_experience_dummies,education_level_dummies,major_discipline_level_dummies],axis ="columns")
```

```python
df6
```

```python
df7 = df6.drop(["relevent_experience","education_level","major_discipline"], axis = "columns")
```

```python
df7
```

# Model Building

```python
x = df7.drop("target",axis = "columns")
```

```python
y = df7["target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth = 8,min_samples_leaf = 9,max_features = "sqrt")
```

```python
dtc.fit(x,y)
```

```python
dtc.predict(x_test)
```

```python
dtc.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
```

```python
cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state = 0)
cross_val_score(dtc, x, y, cv= cv)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 4,max_leaf_nodes = 30)
```

```python
rfc.fit(x,y)
```

```python
rfc.score(x_test,y_test)
```

```python
cv = ShuffleSplit(n_splits=5,test_size=0.3, random_state= 10)
cross_val_score(rfc, x, y, cv= cv)
```
